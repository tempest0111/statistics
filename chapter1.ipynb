```python
from urllib.parse import urljoin
base_url = "https://raw.githubusercontent.com/tempest0111/statistics/main/csv/"
ex1_1 = urljoin(base_url, 'ex1-1.csv')
print(ex1_1)
```

```python
import pandas as pd 

# csvファイルの読み込み
data = pd.read_csv(ex1_1)

# 先頭5行を表示
data.head()
```

```python
# 平均値を算出
print('平均値')
print(data.mean())

# 標準偏差を算出
print('標準偏差')
print(data.std())
```

```python
# 基本統計量を算出
# 小数点第2位で四捨五入
data.describe().round(1)
```
